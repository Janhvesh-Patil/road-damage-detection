```python
from ultralytics import YOLO
import os
```

```python
"""
Expected dataset structure:

dataset/
├── train/
│   ├── images/
│   └── labels/
├── val/
│   ├── images/
│   └── labels/
"""

print("Ensure dataset is placed in the above structure before training.")
```

```python
data_yaml = """
path: ./dataset
train: train/images
val: val/images

names:
  0: Longitudinal crack
  1: Transverse crack
  2: Alligator crack
  3: Other corruption
  4: Pothole
"""

with open("data.yaml", "w") as f:
    f.write(data_yaml)

print("data.yaml created successfully.")
```

```python
model = YOLO("yolov8n.pt")
print("YOLOv8n model loaded.")
```

```python
model.train(
    data="data.yaml",
    epochs=15,
    imgsz=512,
    batch=16,
    workers=2,
    pretrained=True,
    name="road_damage_yolov8n"
)
```

```python
metrics = model.val()
print(metrics)
```
